# Use Hierarchical clustering to cluster the digit’s dataset a. Agglomerate b. Divisive


```python
#generate a sample dataset of 50 samples 
X, y = make_blobs(n_samples=50, centers=3, random_state=42)
```

```python
# Perform agglomerative clustering
linkage_matrix = linkage(X, method='ward')
dendrogram(linkage_matrix)
plt.title("Agglomerative Clustering Dendrogram")
plt.show()
```

```python
# # Perform divisive clustering
# linkage_matrix = linkage(X.T, method='ward')
# dendrogram(linkage_matrix)
# plt.title("Divisive Clustering Dendrogram")
# plt.show()




def divisive_clustering(X, n_clusters):
    model = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
    labels = model.fit_predict(X)
    if n_clusters == 1:
        return [labels]
    else:
        clusters = []
        for cluster_id in range(n_clusters):
            indices = np.where(labels == cluster_id)[0]
            if len(indices) == 1:
                clusters.append([labels[indices[0]]])
            else:
                subclusters = divisive_clustering(X[indices], 1)
                for subcluster in subclusters:
                    clusters.append([label + len(clusters) for label in subcluster])
        return clusters

clusters = divisive_clustering(X, 3)

# Plot the clusters
plt.figure(figsize=(8, 6))
colors = ['blue', 'green', 'red']
for cluster_id, cluster in enumerate(clusters):
    for index in cluster:
        plt.scatter(X[index][0], X[index][1], color=colors[cluster_id])
plt.title("Divisive Clustering")
plt.show()
```
